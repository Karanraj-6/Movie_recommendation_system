```python
import numpy as np 
import pandas as pd
```

```python
df_standard1=pd.read_csv('TMDB_movies_Hindi.csv')
df_standard2=pd.read_csv('TMDB.csv')
```

```python
# column that going to use

df=df_standard[['movie_id','title','overview','genres','production_companies','cast', 'crew']]
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df=df.drop_duplicates()
df.shape
```

```python
import ast
```

```python
def fetch_genres(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
df['genres']=df['genres'].apply(fetch_genres)
```

```python
def fetch_production_comapanies(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
df['production_companies']=df['production_companies'].apply(fetch_production_comapanies)
```

```python
def fetch_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter != 3:
            l.append(i['original_name'])
            counter+=1
        else:
            break
    return l
```

```python
df['cast']=df['cast'].apply(fetch_cast)
```

```python
df['crew'][7].values
```

```python
import ast
import pandas as pd

def fetch_director(text):
    directors = []
    try:
        # Safely evaluate the string to a Python object
        data_list = ast.literal_eval(text)
        
        # Check if the evaluated object is a list
        if isinstance(data_list, list):
            for item in data_list:
                if item.get('department') == 'Directing':
                    directors.append(item.get('original_name'))
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing input: {e}")
    return directors
```

```python
# Apply the fetch_director function to create a new column with the directors
df['crew'] = df['crew'].apply(fetch_director)

# Delete rows where the directors list is empty
df = df[df['crew'].apply(lambda x: isinstance(x, list) and len(x) > 0)]
```

```python
df
```

```python
import re

def remove_noise(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = re.sub(r'\s+', '', text)
    text = text.strip()
    return text
```

```python
df['overview']=df['overview'].apply(remove_noise)
```

```python
df['overview']=df['overview'].apply(lambda X:X.split( ))
```

```python
df['data']=df['cast']+df['crew']+df['genres']+df['overview']+df['production_companies']
```

```python
new_df=df[['movie_id','title','data']]
```

```python
new_df
```

```python
new_df=new_df.reset_index()
new_df=new_df.drop('index',axis=1)
```

```python

def remove_whitespaces(text):
    return [re.sub(r'\s+', '', word) for word in text]

new_df['data']=new_df['data'].apply(remove_whitespaces)
```

```python
new_df
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

new_df['data'] = new_df['data'].apply(lambda text: [lemmatizer.lemmatize(x.lower()) for x in text])
```

```python
new_df
```

```python
new_df['data'] = new_df['data'].apply(lambda x: ' '.join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(stop_words='english')

features_data=cv.fit_transform(new_df['data'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
simularity=cosine_similarity(features_data)
simularity
```
